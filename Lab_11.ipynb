# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
def tsp_with_path(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]
    path_store = [[-1]*(1<<n) for _ in range(n)] # To store the next city in the optimal path

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0] # Return to start city

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city in range(n):
            # If next_city is not visited (i.e., its bit is not set in mask)
            if (mask >> next_city) & 1 == 0:
                # Calculate new cost
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))

                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city

        dp[city][mask] = ans
        path_store[city][mask] = best_next_city
        return ans

    # Start TSP from city 0, with only city 0 visited
    min_cost = visit(0, 1)

    # Reconstruct the path
    current_path = [0]
    current_city = 0
    current_mask = 1 # Mask for only city 0 visited

    while True:
        next_city = path_store[current_city][current_mask]
        if next_city == -1:
            break # Should ideally not happen until last city is processed

        current_path.append(next_city)
        current_city = next_city
        current_mask |= (1 << current_city) # Add next_city to the mask

        if current_mask == (1 << n) - 1: # All cities visited
            current_path.append(0) # Return to the starting city
            break

    print("Path:", current_path)
    return min_cost

# Example usage with the existing graph:
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

print("Minimum travelling cost (with path):", tsp_with_path(graph))
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
# Function to get graph input from the user
def get_graph_from_user():
    while True:
        try:
            n = int(input("Enter the number of cities: "))
            if n <= 0:
                print("Number of cities must be positive. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    graph = []
    print("Enter the adjacency matrix row by row. Enter 'inf' for no direct path.")
    print("Example for a 3-city graph:")
    print("0 10 15")
    print("10 0 20")
    print("15 20 0")

    for i in range(n):
        while True:
            try:
                row_str = input(f"Enter row {i} (space-separated costs): ")
                row_values = row_str.split()
                if len(row_values) != n:
                    print(f"Expected {n} values, but got {len(row_values)}. Please re-enter.")
                    continue
                current_row = []
                for val_str in row_values:
                    if val_str.lower() == 'inf':
                        current_row.append(float('inf'))
                    else:
                        current_row.append(int(val_str))
                graph.append(current_row)
                break
            except ValueError:
                print("Invalid input. Costs must be integers or 'inf'. Please re-enter.")

    return graph

# Get graph from user
user_graph = get_graph_from_user()

# Now, call the tsp_with_path function with the user-provided graph
print("\nCalculating TSP for the user-provided graph...")
min_cost = tsp_with_path(user_graph)
print("Minimum travelling cost (with path) for your graph:", min_cost)
```
